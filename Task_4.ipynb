###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
dep_grp=df3.groupby('Department',sort=False)['Previous Geekions'].max()
name1=df3['Name'].where(df3['Previous Geekions']==dep_grp[0])
name2=df3['Name'].where(df3['Previous Geekions']==dep_grp[1])
name3=df3['Name'].where(df3['Previous Geekions']==dep_grp[2])
name1=list(name1[0])[0]
name2=list(name2[3])[1]
name3=list(name3[0])[1]
score1=(dep_grp[0])
score2=(dep_grp[1])
score3=(dep_grp[2])
name_list=[name1,name2,name3]
score_list=[score1,score2,score3]
department_list=list(dep_grp.keys())
plt.bar(department_list[0],score_list[0],width=0.1,tick_label=name_list[0])
plt.bar(department_list[1],score_list[1],width=0.1,tick_label=name_list[1])
plt.bar(department_list[2],score_list[2],width=0.1,tick_label=name_list[2])
plt.xticks(range(0,3),name_list,fontsize=10)
plt.legend(department_list)
plt.xlabel("Toppers")
plt.ylabel("Geekoins")
plt.title('Department wise toppers',fontsize=25,color='red')
plt.ylim(7000,28000)
for i in range(len(score_list)):
  plt.text(department_list[i],score_list[i],score_list[i])
plt.show()
print('*'*80)
plt.plot(score_list,department_list,marker='o')
plt.xlabel("Geekoins")
plt.ylabel("Department")
plt.title("Department wise toppers",fontsize=25,color='red')
for i, label in enumerate(name_list):
    plt.annotate(label, (score_list[i], department_list[i]))
plt.xlim(7000,28000)
plt.show()
print('*'*80)
plt.barh(department_list[0],score_list[0])
plt.barh(department_list[1],score_list[1])
plt.barh(department_list[2],score_list[2])
plt.xlabel("Geekoins")
plt.ylabel("Department")
plt.title("Department wise toppers",fontsize=25,color='red')
for i, label in enumerate(score_list):
    plt.annotate(label, (score_list[i], department_list[i]))
plt.xlim(7000,28000)
plt.legend(name_list,loc='right')
plt.show()
```
